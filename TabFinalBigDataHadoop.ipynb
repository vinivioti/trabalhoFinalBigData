```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax


# Total de acidentes com vítima por bairro em acidentes com embriaguez;


dF1 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/totalVitimas.csv", sep=',', header=0, index_col = None ,encoding='latin_1')

dF2 = dF1[['BAIRRO', 'ACIDENTES']].groupby('BAIRRO', as_index=False).sum()

ax =dF2.plot.barh(x='BAIRRO',y='ACIDENTES', figsize=(30,60), color="red")

ax.set_title('Total de acidentes com vítima por bairro em acidentes com embriaguez',fontsize = 30,color='red')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8, fontsize = 15)

```

```python
import pandas as pd
import matplotlib.axes as ax
# Total de acidentes por tipo de pavimento e condição do tempo;


dF2 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/envolvidosTempo.csv", sep=',', header=0, index_col = None ,encoding='latin_1')

dF3 = dF2[['CONDTEMPO', 'PAVIMENTO', 'ACIDENTES']].groupby(['PAVIMENTO','CONDTEMPO'], as_index=True).sum()

ax =dF3.plot.barh(y='ACIDENTES', figsize=(15,15), colormap='inferno')

ax.set_title('Total de acidentes por tipo de pavimento e condição do tempo',fontsize = 20,color='black')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8, fontsize = 15)
```

```python
import pandas as pd
import matplotlib.axes as ax
# Total de pessoas acidentadas por tipo de veiculo e tipo de pavimentação;

dF3 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/tipoVeiculo.csv", sep=',', header=0, index_col = None ,encoding='latin_1')

dF4 = dF3[['TIPOVEICULO', 'TIPOPAVIMENTO', 'ACIDENTES']].groupby(['TIPOVEICULO','TIPOPAVIMENTO'], as_index=True).sum()

ax =dF4.plot.barh(y='ACIDENTES', figsize=(15,15), colormap='summer')

ax.set_title('Total de acidentes por tipo de pavimento e condição do tempo',fontsize = 20,color='green')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8, fontsize = 15)
```

```python
import pandas as pd
# Média de idade dos condutores por tipo de veículo e tipo de acidente;

dF5 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/mediaCondutores.csv", sep=',', header=0, index_col = None ,encoding='latin_1')

df6 = dF5[['especie_veiculo',' tipo_acidente',' media_idade']].groupby(['especie_veiculo',' tipo_acidente'], as_index=True).sum()

ax = df6.plot.barh(y=' media_idade', figsize=(30,60), fontsize = 15, colormap = 'coolwarm')

ax.set_title('Média de idade dos condutores por tipo de veículo e tipo de acidente',fontsize = 30,color='blue')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8,fontsize = 15)
```

```python
import pandas as pd
# Média de idade dos condutores por indicativo de embriaguez;

dF7 = pd.read_csv("drive/MyDrive/TrabalhoFinalBigData/mediaEmbreagues.csv", sep=',', header=0, index_col = None ,encoding='latin_1')

dF8 = dF7[['Embreagues', 'media_idade_por_embreagues']].groupby('Embreagues', as_index=True).sum()

ax =dF8.plot.pie(y='media_idade_por_embreagues', 
                figsize=(10,10), 
                colors = ["#20257c", "#6ce2cb", "#3b7f5b"],
                startangle=180,
                fontsize=18,
                x='Embriagues',explode = [.03, .15, .03], )


ax.set_title('Média de idade dos condutores por indicativo de embriaguez',fontsize = 20,color='purple')

ax.legend(bbox_to_anchor=(1,1), loc=0, borderaxespad=0.8, fontsize = 15)
```
